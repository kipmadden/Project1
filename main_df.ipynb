```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("../references/2019 County Health Rankings Texas Data - v1_0.xlsx")
xls = pd.ExcelFile("../references/2019 County Health Rankings Texas Data - v1_0.xlsx")
xls.sheet_names
```

```python
df1 = pd.read_excel(xls, 'Outcomes & Factors Rankings', header=[0,1])
df1.columns = df1.columns.map('_'.join)
df2 = pd.read_excel(xls, 'Outcomes & Factors SubRankings', header=[0,1])
df2.columns = df2.columns.map('_'.join)
df3 = pd.read_excel(xls, 'Ranked Measure Data', header=[0,1])
df3.columns = df3.columns.map('_'.join)
```

```python
df1.head()
```

```python
#Merge the worksheets to a single DF
main_df=df1.merge(df2).merge(df3)
#Saving data to csv
main_df.to_csv("../output_data/main_df.csv")
main_df.head()
```
